# Part1 - Scrape twitter user handle and store in json format

Important libraries are imported

```python
import json
import pandas as pd
from tweepy import OAuthHandler, API, Cursor
import numpy as np
import matplotlib.pyplot as plt
```

Twitter app is created on dev.twitter.com/ 

The consumer kay, consumer_secret_key, access_token and access_token_secret is copied and poasted from your app

```python
CONSUMER_KEY = 'CONSUMER-KEY'
CONSUMER_SECRET = 'CONSUMER-SECRET-KEY'
ACCESS_TOKEN = 'ACCESS-TOKEN'
ACCESS_TOKEN_SECRET = 'SECRET-ACCESS-TOKEN'
```

OAuthHandler object is created using the consumer keys.

Authorisation access token is set using the access tokens.

Authorisation API object is created using the OAuthHandler.

```python
authorisation = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
authorisation.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
access = API(authorisation)
```

Get user object using the authorisation API object and user account name

```python
user_account = 'midasIIITD'
user = access.get_user(user_account)
```

# Traverse the user details

```python
data_dict = {}
data_dict['name'] = user.name
data_dict['screen_name'] = user.screen_name
data_dict['description'] = user.description
data_dict['statuses_count'] = user.statuses_count
data_dict['friends_count'] = user.friends_count
data_dict['followers_count'] = user.followers_count
```

## Append all the user information in a list

```python
data_tweet = []
for temp in Cursor(access.user_timeline, id=user_account, include_entities=True).items():
    images = 0
    if 'media' in temp.entities:
        images = len(temp.entities['media'])
        
    data_tweet.append({
        'text': temp.text,
        'created_date': str(temp.created_at.date()),
        'created_time': str(temp.created_at.time()),
        'favorite_count': temp.favorite_count,
        'retweet_count': temp.retweet_count,
        'num_images': images
    })
data_dict['tweets'] = data_tweet
```

# Store in a json file

```python
with open('data.json', 'w') as f:
    json.dump(data_dict, f)
```

# Part2 - Parse json file to display data in tabular format

```python
with open('data.json') as f:
    stored = json.load(f)
```
